```python
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
dataset
```

```python
dataset = pd.get_dummies (dataset, drop_first = True , dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset [['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent = dataset [['Profit']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent, dependent,test_size=0.30, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor = regressor.fit(x_train,y_train)
```

```python
#MSE
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion="squared_error", max_features=None, splitter="random", random_state=0)
regressor = regressor.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion="squared_error", max_features=None, splitter="best", random_state=0) 
regressor = regressor.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion="squared_error", max_features="sqrt", splitter="random", random_state=0) # best & random
regressor = regressor.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion="squared_error", max_features="log2", splitter="random", random_state=0) # best & random
regressor = regressor.fit(x_train,y_train)
```

```python
#mae
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion="absolute_error", max_features=None, splitter="random", random_state=0) # best & random
regressor = regressor.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion="absolute_error", max_features="sqrt", splitter="random", random_state=0) # best & random
regressor = regressor.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion="absolute_error", max_features="log2", splitter="random", random_state=0) # best & random
regressor = regressor.fit(x_train,y_train)
```

```python
# friedman_mse, auto,sqrt,log2
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion="friedman_mse", max_features=None, splitter="random", random_state=0) # best & random
regressor = regressor.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion="friedman_mse", max_features="sqrt", splitter="random") # best & random
regressor = regressor.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion="friedman_mse", max_features="log2", splitter="random") # best & random
regressor = regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_socre = r2_score (y_test,y_pred)
r_socre
```
